# Problem 1
Bir YetGenli yeni aldığı bahçesinin etrafına çit çevirmek istiyor. Bahçenin etrafını çit ile çevirmek için gerekli çit uzunluğunu ve bahçenin alanını hesaplayan fonksiyonu yazınız.

````python
def bahce_cit_uzunlugu(genislik, uzunluk):
    cevre = 2 * (genislik + uzunluk)
    return cevre

def bahce_alani(genislik, uzunluk):
    alan = genislik * uzunluk
    return alan

genislik = 5
uzunluk = 10

cit_uzunlugu = bahce_cit_uzunlugu(genislik, uzunluk)
bahce_alani = bahce_alani(genislik, uzunluk)

print("Bahçenin etrafına çevrilecek çitin uzunluğu: ", cit_uzunlugu)
print("Bahçenin alanı: ", bahce_alani)
````

# Problem 2
Kendisine gönderilen 2 sayı arasındaki tüm asal sayıları bulan fonksiyonu yazınız.

````python
def asal_sayilari_bul(sayi1, sayi2):
    asal_sayilar = []

    for sayi in range(sayi1, sayi2+1):
        if sayi > 1:
            for i in range(2, sayi):
                if (sayi % i) == 0:
                    break
            else:
                asal_sayilar.append(sayi)

    return asal_sayilar

sayi1 = 10
sayi2 = 50

asal_sayilar = asal_sayilari_bul(sayi1, sayi2)

print(f"{sayi1} ile {sayi2} arasındaki asal sayılar: {asal_sayilar}")
````

# Problem 7
List comprehension kullanarak 1'den 100'e kadar olan sayılardan sadece çift sayıları bir listeye atmayı yapmayı çalışın.

````python
def mukemmelSayi(sayi):
    toplam = 0
    for i in range(1, sayi):
        if sayi % i == 0:
            toplam += i
    return toplam == sayi

for sayi in range(1, 1001):
    if mukemmelSayi(sayi):
        print(sayi)
````
