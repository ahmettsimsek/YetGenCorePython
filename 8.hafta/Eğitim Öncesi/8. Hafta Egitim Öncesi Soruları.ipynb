# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def urun_kaydet():
    urun_ad = input("Ürün adı: ")
    urun_fiyat = float(input("Ürün fiyatı: "))
    
    urun_bilgisi = {"ad": urun_ad, "fiyat": urun_fiyat}
    
    with open("urunler.txt", "a") as dosya:
        dosya.write(str(urun_bilgisi) + "\n")
    
    print("Ürün kaydedildi.")
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage

class Bus(Vehicle):
    pass

bus = Bus("School Volvo", 180, 12)
print("Name:", bus.name, "Speed:", bus.speed, "Mileage:", bus.mileage)
````
